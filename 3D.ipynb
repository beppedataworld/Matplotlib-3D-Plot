### Part 1: Libraries,magic commands and basic structure for 3D plots

```python
import matplotlib.pyplot as plt
import numpy as np
# from mpl_toolkits.mplot3d import Axes3D   
%matplotlib widget 
import warnings
warnings.filterwarnings("ignore")
```

```python
# plt.axes(projection='3d')
ax=plt.figure().add_subplot(projection='3d')
x=np.arange(-10,10,0.5)
y=np.random.randint(-10,10,40)
z=x**2 + y**2
ax.scatter(x,y,z,c=x,cmap='plasma')
plt.show()
```

### PART 2: 3D Planes: "Zs" & "Zdir" - A simple data analysis example

```python
ax=plt.figure().add_subplot(projection='3d')
plane=[2,4,6,8]
drunk=["<15","15","16-30","31-50",">50"]
europe=['Norway','Malta','Estonia','Wales']
colo=['r','green','b','yellow']
for p,c in zip(plane,colo):
    x=np.random.randint(1,5,20)
    y=np.arange(1,21,1)
    ax.bar(x,y,zs=p,zdir='y',color=c)
ax.set_xlabel('Drunk Ages')
ax.set_ylabel('countries')
ax.set_zlabel('Frequency of drunk people')
# Set the y-ticks to be at the values in the plane list
ax.set_yticks(plane)
# Set the y-tick labels to be the corresponding names in the europe list
ax.set_yticklabels(europe)
ax.set_xticklabels(drunk)

plt.show()
```

### Part III: Surface,Wireframe and Meshgrid

```python
x=np.arange(-5,5,0.1)
y=np.arange(-5,5,0.1)
x2,y2=np.meshgrid(x,y)
z=np.sin(x2)+np.cos(y2)
f=plt.figure()
axx=f.add_subplot(projection='3d')
axx.plot_surface(x2,y2,z,rstride=30,cstride=30)
```
